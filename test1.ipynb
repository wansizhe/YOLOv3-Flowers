```python
import numpy as np
import pandas as pd
```

```python
labels = pd.read_csv('label.csv')
```

```python
labels.head()
```

```python
labels['image'] += 1
```

```python
labels.head()
```

```python
len(labels)
```

```python
group = []
temp = []
start = 1
end = 1
last = 77
for i in range(len(labels)):
    if i == 0:
        continue
    if labels['label'][i] == last:
        end += 1
    else:
        temp.append(start)
        temp.append(end)
        temp.append(last)
        temp.append(end-start+1)
        group.append(temp)
        temp = []
        last = labels['label'][i]
        end += 1
        start = end
```

```python
len(group)
```

```python
dt = pd.DataFrame(group, columns=['start', 'end', 'id', 'len'])
```

```python
dt
```

```python
dt.to_csv('group.csv')
```
